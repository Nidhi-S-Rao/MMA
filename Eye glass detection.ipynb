```python
from PIL import Image
import cv2
import numpy as np
import requests
```

```python
# Reading image form url
image = Image.open('eyeglass.jpg')
image = image.resize((450,250))
image_arr = np.array(image)
image
```

```python
grey = cv2.cvtColor(image_arr,cv2.COLOR_BGR2GRAY)
Image.fromarray(grey)
```

```python
eye_cascade_src = 'haarcascade_eye_tree_eyeglasses.xml'
eye_cascade = cv2.CascadeClassifier(eye_cascade_src)
eye = eye_cascade.detectMultiScale(grey, 1.1, 2)
```

```python
for (x,y,w,h) in eye:
  cv2.rectangle(image_arr,(x,y),(x+w,y+h),(255,0,0),2)
Image.fromarray(image_arr)
```
