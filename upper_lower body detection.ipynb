```python
from PIL import Image
import cv2
import numpy as np
import requests
```

```python
# Reading image form url
image = Image.open('people.jpg')
image = image.resize((450,250))
image_arr = np.array(image)
image
```

```python
grey = cv2.cvtColor(image_arr,cv2.COLOR_BGR2GRAY)
Image.fromarray(grey)
```

```python
people_cascade_src = 'haarcascade_upperbody.xml'
people_cascade = cv2.CascadeClassifier(people_cascade_src)
people = people_cascade.detectMultiScale(grey, 1.1, 2)
```

```python
for (x,y,w,h) in people:
  cv2.rectangle(image_arr,(x,y),(x+w,y+h),(255,0,0),2)
Image.fromarray(image_arr)
```

```python
people_cascade_src = 'haarcascade_lowerbody.xml'
people_cascade = cv2.CascadeClassifier(people_cascade_src)
people = people_cascade.detectMultiScale(grey, 1.1, 2)
```

```python
for (x,y,w,h) in people:
  cv2.rectangle(image_arr,(x,y),(x+w,y+h),(255,0,0),2)
Image.fromarray(image_arr)
```
