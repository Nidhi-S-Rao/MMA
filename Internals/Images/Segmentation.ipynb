```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('mahe.jpeg')
```

```python
cv2_imshow(img)
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
cv2_imshow(hsv)
```

```python
import numpy as np
lower_orange = np.array([0, 100, 100])
upper_orange = np.array([15, 255, 255])

```

```python

mask = cv2.inRange(hsv, lower_orange, upper_orange)
cv2_imshow(mask)
```

```python
orange_object = cv2.bitwise_and(img, img, mask=mask)
```

```python
cv2_imshow(orange_object)
```

```python
cv2.bitwise_not(orange_object)
```

```python
img=cv2.imread('ipl.jpg')
```

```python
cv2_imshow(img)
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
cv2_imshow(hsv)
```

```python
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([140, 255, 255])


mask = cv2.inRange(hsv, lower_blue, upper_blue)


blue_object = cv2.bitwise_and(img, img, mask=mask)
```

```python
cv2_imshow(blue_object)
```

```python
cv2.bitwise_not(blue_object)
```
