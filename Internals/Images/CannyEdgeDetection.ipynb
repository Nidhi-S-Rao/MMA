Applying Canny edge detection to extract only the edges in the images

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('cubes.png')
```

```python
cv2_imshow(img)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray)
```

```python
filter=cv2.GaussianBlur(gray,(3,3),0)
```

```python
cv2_imshow(filter)
```

```python
cv2.Canny(gray,60,100)
```
